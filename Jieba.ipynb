```python
split = []
All = []
u=[]
stopwords=[]
import csv
import jieba
import pandas as pd

df=pd.read_csv(r'財報狗all.csv',encoding='utf_8_sig')

df["斷詞"]=None

x = 0
for r in df["內容"]:
    All.append(df["內容"][x])
    x+=1
for stop in open(r'stopwords.txt',encoding='utf_8_sig'):
        stopwords.append(stop.strip())  
num = 0        
for row in All:
    cut = jieba.lcut(row, cut_all=False)
    for word in cut:
        if word not in stopwords:
            split.append(word)
            u.append(word)
    df.at[num,'斷詞'] = split
    num+=1
    split=[]  
df.to_csv('財報狗斷詞.csv', index=False,encoding='utf_8_sig')                
```
