```python
#爬取文章內容   
content = []
for post in postlist:
    content.append(post.text)    
    
#爬取文章時間
o1 = []
o2 = []
for oc1 in date:
    o1.append(str(oc1))
for oc2 in o1:
    o2.append(re.findall(r'title="(\d*/\d*/\d* \d*:\d*)',oc2))
n=0
o3 = []
time = []
for oc3 in o2:
    o3.append(str(o2[n]).replace("['",""))
    time.append(str(o3[n]).replace("']",""))
    n+=1

#爬取文章標題
t1 = []
t2 = []
for tc1 in topic:
    t1.append(str(tc1))
for tc2 in t1:
    t2.append(re.findall(r'title="(\w+)',tc2))    
tnum = t2[:len(content)]
title = []
for tc3 in tnum:
    title.append(str(tc3).replace("', '')","").replace("('","").replace("[","").replace("]","").replace("'",""))  
```
